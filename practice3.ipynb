```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('NetflixDataset.csv',encoding='latin-1')
```

```python
df
```

```python
Language=df.Languages.str.get_dummies(",")
```

```python
lang=Language.columns.str.strip().values.tolist()
```

```python
len(lang)
```

```python
lang_count=dict()
for i in lang:
    p=df.Languages.str.count(i).sum()
    lang_count[i]=p
```

```python
df1=pd.DataFrame(lang_count,index=[1])
```

```python
df1
```

```python
df1_transposed = df1.T
```

```python
df1_transposed['index']=df1_transposed.index
```

```python
df2=df1_transposed.rename(columns={1:'counts','index':"language"})
```

```python
df2
```

```python
df2.reset_index(drop=True,inplace=True)
```
